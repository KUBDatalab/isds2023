```python
import pandas as pd
import matplotlib.pyplot as plt
pd.v
```

```python
data = pd.read_csv("fitted_models/all_pred_lasso_year.csv")
df = data.query("muni_code == 540").copy()
```

```python
ejerlejlighed_df = df[df['housing_type'] == 'Ejerlejlighed']
rækkehus_df = df[df['housing_type'] == 'Rækkehus']
villa_df = df[df['housing_type'] == 'Villa']

# Lav plot for Ejerlejlighed
plt.plot(ejerlejlighed_df['year'], ejerlejlighed_df['y'], label='Ejerlejlighed y', linestyle='-', color = "red")
plt.plot(ejerlejlighed_df['year'], ejerlejlighed_df['y_pred'], label='Ejerlejlighed y_pred', linestyle='--', color = "red")

# Lav plot for Rækkehus
plt.plot(rækkehus_df['year'], rækkehus_df['y'], label='Rækkehus y', linestyle='-', color = "blue")
plt.plot(rækkehus_df['year'], rækkehus_df['y_pred'], label='Rækkehus y_pred', linestyle='--', color = "blue")

# Lav plot for Villa
plt.plot(villa_df['year'], villa_df['y'], label='Villa y', linestyle='-', color = "green")
plt.plot(villa_df['year'], villa_df['y_pred'], label='Villa y_pred', linestyle='--', color = "green")

# Tilføj labels og titel
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Plot af boligtyper')
plt.legend(loc="upper left", bbox_to_anchor=(1,1))

plt.show()
```
