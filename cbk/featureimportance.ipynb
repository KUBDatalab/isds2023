```python
from pathlib import Path
import pandas as pd
```

```python
fp = Path("../ml_data/fitted_models/")
files = list(fp.glob('coefficients*.csv'))


elastic_features ='../ml_data/fitted_models/coefficients_elastic.csv'
lasso_features = '../ml_data/fitted_models/coefficients_lasso.csv'
ridge_features = '../ml_data/fitted_models/coefficients_ridge.csv'
```

```python
lasso_features = pd.read_csv(lasso_features)
```

```python
lasso_imp_features = pd.read_csv(lasso_features)\
                    .query("value != 0")\
                    .groupby(["parameter"])\
                    .agg("count")\
                    .reset_index()\
                    .sort_values("value", ascending = False)\
                    .head(10)

ridge_imp_features = pd.read_csv(ridge_features)\
                    .query("value != 0")\
                    .groupby(["parameter"])\
                    .agg("count")\
                    .reset_index()\
                    .sort_values("value", ascending = False)\
                    .head(10)
elastic_imp_features = pd.read_csv(elastic_features)\
                    .query("value != 0")\
                    .groupby(["parameter"])\
                    .agg("count")\
                    .reset_index()\
                    .sort_values("value", ascending = False)\
                    .head(10)
```
