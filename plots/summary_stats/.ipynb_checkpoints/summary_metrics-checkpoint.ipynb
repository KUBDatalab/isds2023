```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pd.set_option('display.max_rows', None)
```

```python
def format_min_max(group):
    min_rmse, max_rmse = round(group['rmse'].min(), 2), round(group['rmse'].max(), 2)
    min_r2, max_r2 = round(group['r2'].min(), 2), round(group['r2'].max(), 2)
    return pd.Series({
        'rmse range': f"{min_rmse} - {max_rmse}",
        'r2 range': f"{min_r2} - {max_r2}"
    })

pd.concat([pd.read_csv("../../ml_data/fitted_models/metrics_ridge.csv").agg({
    'rmse': ['min', 'max'],
    'r2': ['min', 'max']
    }).assign(regularization = "ridge", cut = "8020"),
pd.read_csv("../../ml_data/fitted_models/metrics_lasso.csv").agg({
    'rmse': ['min', 'max'],
    'r2': ['min', 'max']
    }).assign(regularization = "lasso", cut = "8020") ,
pd.read_csv("../../ml_data/fitted_models/metrics_elastic.csv").agg({
    'rmse': ['min', 'max'],
    'r2': ['min', 'max']
    }).assign(regularization = "elastic", cut = "8020") ,
pd.read_csv("../../ml_data/fitted_models/all_metrics_elastic_year.csv").agg({
    'rmse': ['min', 'max'],
    'r2': ['min', 'max']
    }).assign(regularization = "elastic", cut = "TZ") ,
pd.read_csv("../../ml_data/fitted_models/all_metrics_lasso_year.csv").agg({
    'rmse': ['min', 'max'],
    'r2': ['min', 'max']
    }).assign(regularization = "lasso", cut = "TZ"), 
pd.read_csv("../../ml_data/fitted_models/all_metrics_ridge_year.csv").agg({
    'rmse': ['min', 'max'],
    'r2': ['min', 'max']
    }).assign(regularization = "ridge", cut = "TZ") ]).groupby(['regularization', 'cut'])\
          .apply(format_min_max)\
          .reset_index().sort_values(["cut", "regularization"])


```

```python
data = pd.read_csv("../../ml_data/fitted_models/all_pred_lasso_year.csv") 

df = data.query("muni_code == 615").copy() 

# Prediction - “tidsserierne” . Plots på udvalgte kommuner 

ejerlejlighed_df = df[df['housing_type'] == 'Ejerlejlighed'] 
rækkehus_df = df[df['housing_type'] == 'Rækkehus'] 
villa_df = df[df['housing_type'] == 'Villa'] 

# make plots for condomoniums
plt.plot(ejerlejlighed_df['year'], ejerlejlighed_df['y'], label='Condomonium', linestyle='-', color = "red") 
plt.plot(ejerlejlighed_df['year'], ejerlejlighed_df['y_pred'], label='Condomonium predicted', linestyle='--', color = "red") 

# make plots for townhouses
plt.plot(rækkehus_df['year'], rækkehus_df['y'], label='Townhouse', linestyle='-', color = "blue") 
plt.plot(rækkehus_df['year'], rækkehus_df['y_pred'], label='Townhouse predicted', linestyle='--', color = "blue") 

# make plots for detached homes
plt.plot(villa_df['year'], villa_df['y'], label='Detached home', linestyle='-', color = "green") 
plt.plot(villa_df['year'], villa_df['y_pred'], label='Detached home predicted', linestyle='--', color = "green") 

# Add labels and titles
plt.xlabel('Year') 
plt.ylabel('Value scaled') 

plt.title('Actual and predicted square meter prices for Horsens\n as fitted with LASSO regularization ') 
plt.legend(loc="upper left", bbox_to_anchor=(1,1)) 

plt.axvline(x=2015, color='black', linestyle='--')

# save the plot
plt.savefig('horsens-yearly.png', bbox_inches='tight')
plt.show() 

```

Learning curve on the LASSO explanatory model for Horsens

```python
data = pd.read_csv("../../ml_data/fitted_models/learning_curves_lasso.csv") 
```

```python
sønderborg = data.query("muni_code == 540").copy() 

sønderborg_learn, ax = plt.subplots(figsize=(7,3)) 
ax.plot(sønderborg["sample size"],np.sqrt(sønderborg["Test"]), alpha=0.25, linewidth=2, label ='Test', color='blue') # negated, because we already use neg_MSE 
ax.plot(sønderborg["sample size"],np.sqrt(sønderborg["Train"]), alpha=0.25, linewidth=2, label='Train', color='orange') # negated, because we already use neg_MSE 
ax.set_title('Learning curve for the explanatory model for Sønderborg\n using LASSO regularization') 
ax.set_ylabel('Root-Mean squared error') 
ax.set_xlabel('Number of training samples') 
ax.legend(); 


sønderborg_learn.savefig('sønderborg-learn.png', bbox_inches='tight')
```

```python
horsens = data.query("muni_code == 615").copy() 

horsens_learn, ax = plt.subplots(figsize=(7,3)) 
ax.plot(horsens["sample size"],np.sqrt(horsens["Test"]), alpha=0.25, linewidth=2, label ='Test', color='blue') # negated, because we already use neg_MSE 
ax.plot(horsens["sample size"],np.sqrt(horsens["Train"]), alpha=0.25, linewidth=2, label='Train', color='orange') # negated, because we already use neg_MSE 
ax.set_title('Learning curve for the explanatory model for Horsens\n using LASSO regularization') 
ax.set_ylabel('Root-Mean squared error') 
ax.set_xlabel('Number of training samples') 
ax.legend(); 


horsens_learn.savefig('horsens-learn.png', bbox_inches='tight')
```

Table with all RMSE and R2 for all models and municipalities

```python
# municipal codes and names
municipal_codes = pd.read_csv("../../Feature_data/municipality_number_codes.csv", sep = ";", header = None).rename(columns={0:"Municipality", 1:"muni_code"})

all_metrics = pd.concat([pd.read_csv("../../ml_data/fitted_models/metrics_ridge.csv").assign(regularization = "ridge", split = "80/20"),
                         pd.read_csv("../../ml_data/fitted_models/metrics_lasso.csv").assign(regularization = "lasso", split = "80/20") ,
                         pd.read_csv("../../ml_data/fitted_models/metrics_elastic.csv").assign(regularization = "elastic", split = "8020") ,
                         pd.read_csv("../../ml_data/fitted_models/all_metrics_elastic_year.csv").assign(regularization = "elastic", split = "Temporal") ,
                         pd.read_csv("../../ml_data/fitted_models/all_metrics_lasso_year.csv").assign(regularization = "lasso", split = "Temporal"), 
                         pd.read_csv("../../ml_data/fitted_models/all_metrics_ridge_year.csv").assign(regularization = "ridge", split = "Temporal") ])
all_metrics = all_metrics.merge(municipal_codes, how='left').rename(columns={'rmse':"RMSE", 'r2':'R2'})
all_metrics = all_metrics[['Municipality', 'split', 'regularization', 'RMSE', 'R2']].sort_values(['Municipality', 'split', 'regularization'])
```

```python
all_metrics.to_csv("../../appendices/appendix1.csv")
```
