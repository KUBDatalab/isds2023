```python
import re
import numpy as np
def get_square_meter_price(row): # Get the prices for a square meters, for examle 13158
    strings = re.split('m²', row)[1].replace('.', '')
    strings = ''.join(re.findall('\d', strings))
    try:
        return int(strings)
    except ValueError:
        return np.NaN

all_df['square_meter_price'] = all_df['m² & Kr. / m²'].apply(lambda row: get_square_meter_price(row))
        
        
    

```
