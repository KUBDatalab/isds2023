```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Load data
gini_data_path = Path.cwd() / '../Feature_data' # path of files to be found
gini_data = pd.read_csv(gini_data_path / 'gini_index.csv')

# Load data
unemplo_data_path = Path.cwd() / '../Feature_data' # path of files to be found
unemplo_data = pd.read_csv(unemplo_data_path / 'unemployment_data.csv')
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,ncols=1, figsize=(7, 4))

sns.lineplot(gini_data, x='year', y='gini_index', ax=ax0)
ax0.set_title('Line Plot from Gini index data')
ax0.set_xlabel('Year')
ax0.set_ylabel('Gini Index')

sns.lineplot(unemplo_data, x='year', y='unemployed', ax=ax1)
ax1.set_title('Line Plot from unemploment data')
ax1.set_xlabel('Year')
ax1.set_ylabel('Unemployed')
ax1.tick_params(axis='y')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
